```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape and create a data frame
def scrape_and_create_dataframe(url, table_class, column_names):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', {'class': table_class})
    rows = table.find_all('tr')
    data = []
    for row in rows[1:]:
        columns = row.find_all('td')
        values = [column.text.strip() for column in columns]
        data.append(values)
    df = pd.DataFrame(data, columns=column_names)
    return df

# Scrape top 10 ODI teams in women's cricket
url_teams = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
table_class_teams = 'table rankings-table'
column_names_teams = ['Position', 'Team', 'Matches', 'Points', 'Rating']
df_teams = scrape_and_create_dataframe(url_teams, table_class_teams, column_names_teams)

# Scrape top 10 women's ODI batting players
url_batting = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting'
table_class_batting = 'table rankings-table'
column_names_batting = ['Position', 'Player', 'Team', 'Rating']
df_batting = scrape_and_create_dataframe(url_batting, table_class_batting, column_names_batting)

# Scrape top 10 women's ODI all-rounders
url_allrounders = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder'
table_class_allrounders = 'table rankings-table'
column_names_allrounders = ['Position', 'Player', 'Team', 'Rating']
df_allrounders = scrape_and_create_dataframe(url_allrounders, table_class_allrounders, column_names_allrounders)

# Display the data frames
print("Top 10 ODI Teams in Women's Cricket")
print(df_teams.head(10))
print()

print("Top 10 Women's ODI Batting Players")
print(df_batting.head(10))
print()

print("Top 10 Women's ODI All-rounders")
print(df_allrounders.head(10))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the web page
url = 'https://www.cnbc.com/world/?region=world'
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the news articles
articles = soup.find_all('div', {'class': 'Card-titleContainer'})

# Extract the headline, time, and news link for each article
headlines = []
times = []
news_links = []

for article in articles:
    headline = article.find('a').text.strip()
    time = article.find('time')['datetime']
    news_link = 'https://www.cnbc.com' + article.find('a')['href']
    
    headlines.append(headline)
    times.append(time)
    news_links.append(news_link)

# Create a data frame using pandas
df = pd.DataFrame({'Headline': headlines, 'Time': times, 'News Link': news_links})

# Display the data frame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the web page
url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the articles
articles = soup.find_all('li', {'class': 'js-article'})

# Extract the details for each article
paper_titles = []
authors_list = []
published_dates = []
paper_urls = []

for article in articles:
    title = article.find('h3', {'class': 'js-article-title'}).text.strip()
    authors = article.find('div', {'class': 'authors'}).text.strip()
    date = article.find('div', {'class': 'published-online'}).text.strip()
    url = article.find('h3', {'class': 'js-article-title'}).a['href']
    
    paper_titles.append(title)
    authors_list.append(authors)
    published_dates.append(date)
    paper_urls.append(url)

# Create a data frame using pandas
df = pd.DataFrame({
    'Paper Title': paper_titles,
    'Authors': authors_list,
    'Published Date': published_dates,
    'Paper URL': paper_urls
})

# Display the data frame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the web page
url = 'https://www.dineout.co.in/delhi-restaurants'
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the restaurant cards
restaurant_cards = soup.find_all('li', {'class': 'restaurant-card'})

# Extract the details for each restaurant
restaurant_names = []
cuisines = []
locations = []
ratings = []
image_urls = []

for card in restaurant_cards:
    name = card.find('div', {'class': 'restnt-info cursor'}).h3.text.strip()
    cuisine = card.find('div', {'class': 'double-line-ellipsis'}).text.strip()
    location = card.find('span', {'class': 'double-line-ellipsis'}).text.strip()
    rating = card.find('span', {'class': 'star-ratings-sprite-rating'}).get('style')
    image_url = card.find('div', {'class': 'restaurant-card__thumb'}).img.get('src')
    
    # Extract only the numeric rating value from the style attribute
    rating_value = float(rating.split(':')[1].strip('%;')) / 20
    
    restaurant_names.append(name)
    cuisines.append(cuisine)
    locations.append(location)
    ratings.append(rating_value)
    image_urls.append(image_url)

# Create a data frame using pandas
df = pd.DataFrame({
    'Restaurant Name': restaurant_names,
    'Cuisine': cuisines,
    'Location': locations,
    'Ratings': ratings,
    'Image URL': image_urls
})

# Display the data frame
print(df)
```
